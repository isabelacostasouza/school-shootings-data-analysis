# Previsão

A previsão majoritária que o trabalho constatou foi que o numero de incidentes em um estado pode ser gerado com uma fidelidade alta a partir do numero de leis abordando tal problematica que o mesmo possui. Quanto mais leis, mais incidentes, mas isso com certeza nao é um efeito direto das leis mas sim provavelmente do numero de incidentes: estados com mais incidentes começaram a implantar mais leis, mas que acabaram nao resolvendo muito o problema de fato.

# Conclusões

A partir dos dados analisados, constata-se a dificuldade e incerteza no que se diz respeito a tiroteios escolares, sendo que leis nao pareceram afetar a ocorrencia dos mesmos de forma significativa e que seus atiradores na maior parte dos casos nao sao pessoas socialmente marginalizadas ou que mostrem padroes preveníveis.
